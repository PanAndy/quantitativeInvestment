```python
from datetime import datetime

import pandas as pd

from vnpy.trader.database import database_manager
from vnpy.trader.constant import Exchange, Interval
```

```python
# 从数据库读取数据
data = database_manager.load_bar_data(
    symbol="IF2004",
    exchange=Exchange.CFFEX,
    interval=Interval.MINUTE,
    start=datetime(2020,1,1),
    end=datetime(2020,4,20)
)
```

```python
# 转换成DataFrame
buf = {
    "datetime": [],
    "open": [],
    "high": [],
    "low": [],
    "close": [],
    "volume": [],
}

for bar in data:
    buf["datetime"].append(bar.datetime)
    buf["open"].append(bar.open_price)
    buf["high"].append(bar.high_price)
    buf["low"].append(bar.low_price)
    buf["close"].append(bar.close_price)
    buf["volume"].append(bar.volume)

df = pd.DataFrame(buf)
```

```python
# 收盘价绘图
df["close"].plot()
```
